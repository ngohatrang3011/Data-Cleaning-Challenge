```python
import pandas as pd
```

Read csv file and drop NaN columns

```python
df = pd.read_csv('data_cleaning_challenge.csv', usecols = [i for i in range(9)])
df.head()
```

Drop NaN rows

```python
df = df.replace('-', 'NaN').dropna(subset=['Row Type'])
df.head(15)
```

Assign column ID

```python
#range(df[new_person].shape[0]) = range(0,5994)
new_person = df['Row Type'].str.contains('first name')
df.loc[new_person, 'ID'] = range(1,5995)
df['ID'].fillna(method='ffill', inplace=True)
df['ID'] = df['ID'].astype(int)
df.head(5)
```

Create First Name, Last Name and Date columns

```python
df.loc[new_person, 'First Name'] = df.loc[new_person, 'Row Type'].str.replace('first name: ', '')
df.loc[new_person, 'Last Name'] = df.loc[new_person, 'Iter Number'].str.replace('last name: ', '')
df.loc[new_person, 'Date'] = df.loc[new_person, 'Power1'].str.replace('date: ', '')
df.loc[:, ['First Name', 'Last Name', 'Date']] = df.loc[:, ['First Name', 'Last Name', 'Date']].ffill()
df = df[(df['Row Type'] != 'first name: Person') & (df['Row Type'] != 'Row Type')]
df.head()
```

Move new columns to front

```python
df = df.set_index(['ID','First Name', 'Last Name', 'Date']).reset_index()
df
```
